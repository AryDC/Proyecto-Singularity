```python
import tensorflow as tf #Libreria para hacer Redes Neuronales
import pandas as pd #Panel Data
import matplotlib.pyplot as plt #Graficas
import numpy as np #Numerical Python
```

```python
celsius = np.linspace(-273.15, 273.15, num=100) #Crear una vector de numeros con las cantidad y rango de numeros dados
#e este vector anterior se le aplica una formula para pasar de Celsius -> Farenheits:
farenheit = np.array([(temperatura*1.8)+32 for temperatura in celsius]) 

#Grafica
vista = pd.DataFrame()
vista["celsius"] = pd.Series(celsius)
vista["farenheits"] = pd.Series(farenheit)
vista.plot()
```

```python
#defino el tipo del modelo y dentro sus capas:
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=100, input_shape=[1]),
    tf.keras.layers.Dense(units=3),
    tf.keras.layers.Dense(units=3),
    tf.keras.layers.Dense(units=1)
])

#Compilacion
model.compile(
    optimizer= tf.keras.optimizers.Adam(0.1),
    loss = "mean_squared_error"
)
```

```python
vueltas = model.fit(celsius, farenheit, epochs = 100, verbose=False)
```

```python
plt.plot(vueltas.history["loss"])
```

```python
resultado = model.predict([100])
print(f"{resultado[0][0]:.2f} Celsius")
```
